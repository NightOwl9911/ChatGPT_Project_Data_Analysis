```python
# Importing our libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Loading our dataset
df = pd.read_csv(r"C:\Users\WIN10\Desktop\ChatGPT_Project\Data\1_DataSet_Health\Healthcare_Investments_and_Hospital_Stay (1).csv")

# Column Normalization
df.columns = df.columns.str.lower()
```

```python
# Get the length of stay in AUS over the years
df_aus = df[df['location'] == 'AUS'].copy()

# Plotting the stay of length in Australia over the years
fig, ax = plt.subplots(figsize=(10,6))
sns.set_style("whitegrid")
plt.grid(False)
sns.despine()
sns.lineplot(data=df_aus, x='time', y='hospital_stay')
ax.set_facecolor('black')
ax.tick_params(color='white', labelcolor='white')
ax.spines['bottom'].set_color('white')
ax.spines['left'].set_color('white')
ax.set_xlabel('Year', color='white', fontsize=12)
ax.set_ylabel('Time of Stay (days)', color='white', fontsize=12)
ax.set_ylim(3, 7)
ax.set_title('Length of Stay over the Years', color='white')

fig.patch.set_facecolor('black')
```

```python
# Plot an scatter plot to get the relationship between MRI units and Length of Stay 
fig, ax = plt.subplots(figsize=(10,6))

# Plotting
sns.scatterplot(data=df_aus, x='mri_units', y='hospital_stay', hue='time', palette='Blues_r')

# Styling
ax.spines['bottom'].set_color('white')
ax.spines['left'].set_color('white')
ax.set_facecolor('black')
ax.tick_params(color='white', labelcolor='white')
ax.set_title('Influence of Quantity of MRI Units in Lenght of Stay', color='white')
ax.set_ylabel('Lenght of Stay', color='white', fontsize=12)
ax.set_xlabel('Cound of MRI Units', color='white', fontsize=12)
ax.legend().set_title('Year')


fig.patch.set_facecolor('black')
plt.show()
```
